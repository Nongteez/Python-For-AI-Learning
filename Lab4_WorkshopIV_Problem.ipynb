1.จงสร้างแบบจำลองเพื่อจัดกลุ่มข้อมูลของดอกไอริส ดังนี้
* สามารถนำเข้าข้อมูล wine data จาก sklearn datasets ได้
* สามารถแปลงข้อมูล wine data ให้เป็นตารางใน DataFrame พร้อมแสดงผลออกทางจอภาพ
* สามารถสร้างแบบจำลองโดยใช้เทคนิค K-means clustering โดยกำหนดให้ k = 3 พร้อมแสดงข้อมูลผลลัพธ์ (wine target/y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
* สามารถแสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ออกทางจอภาพ โดยแกนนอนคือข้อมูลในคอลัมน์แรก แกนตั้งคือข้อมูลในคอลัมน์ที่สอง
*   สามารถแสดง Confusion Matrix ของ y_true และ y_pred พร้อมทั้งประเมินแบบจำลองโดยใช้ Accuracy และแสดงออกทางจอภาพได้


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, accuracy_score

#สามารถนำเข้าข้อมูล wine data จาก sklearn datasets ได้ถูกต้อง
iris = datasets.load_iris()


#สามารถแปลงข้อมูล wine data ให้เป็นตารางใน DataFrame พร้อมแสดงผลออกทางจอภาพ
X = pd.DataFrame(iris.data, columns=['Sepal_Length', 'Sepal_Width',
                                     'Petal_Length', 'Petal_Width'])

print("_iris Data column names_")
print(X)

y_true = iris.target
print("\nTarget or Actual Values")
print(y_true)


#สามารถสร้างแบบจำลองโดยใช้เทคนิค K-means clustering โดยกำหนดให้ k = 3
kmeans = KMeans(n_clusters=3)   # k = 3
y_pred = kmeans.fit_predict(X)


#สามารถแสดงข้อมูลผลลัพธ์ (wine target/y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
print("\nPredicted Values ")
print(y_pred)


#สามารถแสดงแผนภาพกราฟจุดของข้อมูล y_true และ y_pred ออกทางจอภาพ
#โดยแกนนอนคือข้อมูลในคอลัมน์แรก แกนตั้งคือข้อมูลในคอลัมน์ที่สอง และแต่ละกลุ่มให้กำหนดสีของจุดต่างกัน
colors = np.array(['yellow', 'red', 'blue'])


#สามารถแสดง Confusion Matrix ของ y_true และ y_pred ออกทางจอภาพได้

#Plot Iris Target
plt.figure(figsize=(6,4))
plt.scatter(X['Sepal_Length'], X['Petal_Length'], c=colors[y_true])
plt.title("Iris Target (Actual)")
plt.xlabel("Sepal_Length")
plt.ylabel("Petal_Length")
plt.show()

#Plot Predicted Clusters
plt.figure(figsize=(6,4))
plt.scatter(X['Sepal_Length'], X['Petal_Length'], c=colors[y_pred])
plt.title("K-Means Predicted Clusters")
plt.xlabel("Sepal_Length")
plt.ylabel("Petal_Length")
plt.show()


#สามารถประเมินแบบจำลองโดยใช้ Accuracy ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ
cm = confusion_matrix(y_true, y_pred)
acc = accuracy_score(y_true, y_pred)

print("Confusion Matrix")
print(cm)
print("\nAccuracy Score")
print(acc)
```

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

import warnings
warnings.filterwarnings("ignore")

from google.colab import drive
drive.mount('/content/drive')

#สามารถนำเข้าข้อมูลจาก weather1.csv ได้ และแสดงในรูปแบบตารางได้
data = pd.read_csv("/content/drive/MyDrive/AI/PythonLab/import/weather1.csv")
print("ข้อมูลเดิมในรูปแบบตาราง")
print(data)


#สามารถแปลงข้อมูลให้อยู่ในรูปแบบ dataset ได้
dataset = data.values.tolist()
print("\nDataset")
print(dataset)


#สามารถแปลง dataset ให้อยู่ในรูปแบบ Numpy boolean array
te = TransactionEncoder()
te_array = te.fit(dataset).transform(dataset)

print("\nแปลง dataset ให้อยู่ในรูปแบบ Numpy boolean array")
print(te_array)


#สามารถแปลง Numpy boolean array ให้อยู่ใน DataFrame ดังรูปในโจทย์
df_boolean = pd.DataFrame(te_array, columns=te.columns_)

print("\nแปลง Numpy boolean array ให้อยู่ใน DataFrame")
print(df_boolean)


#สามารถแสดงผล frequent_patterns ได้ โดยใช้ฟังก์ชั่น apriori และกำหนด min_support=0.2
frequent_patterns = apriori(df_boolean, min_support=0.2, use_colnames=True)

print("\nFrequent Patterns โดยใช้ฟังก์ชั่น apriori และกำหนด (min_support=0.2)")
print(frequent_patterns)


#สามารถสร้าง และแสดงผล Associate Rules โดยกำหนดตัววัดเป็น 'lift' และกำหนด min_threshold=1.6 ได้
rules = association_rules(frequent_patterns, metric="lift", min_threshold=1.6)

print("\nAssociation Rules โดยกำหนดตัววัดเป็น 'lift' และกำหนด min_threshold=1.6")
print(rules)
```
