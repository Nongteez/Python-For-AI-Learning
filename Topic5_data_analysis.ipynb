**Data Frame**

```python
import pandas as pd
import numpy as np


data = {'id':[1,2,3],'name':['ant','bird','cat'],'age':[20,21,30],'address':['SUT','KKV','SPU']}
data
#data = np.array([(1,2,3),(4,5,6)])

df1 = pd.DataFrame(data)
df1
```

```python
#สร้าง DataFrame 2
data2 = {'id':[4],'name':['dog'],'age':[15]}
df2 = pd.DataFrame(data2)
df2
```

```python
#รวม Dataframe 1 กับ 2 เข้าด้วยกันเปลี่ยนให้เป็น Index เดียวกัน
df = df1._append(df2,ignore_index=True)
df
```

```python
#เพิ่ม DataFrame หรือ Key และ Value เข้าไป
import numpy as np
cash = np.array([10,20,30,40])
df['cash'] = cash
df
```

```python
#แสดง Dataframe จาก แถว Index
#:,0 เอาทุกแถวของ Collumn ที่ 0 ถ้าใส่ -1 จะเอาแถวสุดท้าย

df.iloc[0]
```

```python
data = df.iloc[[0,1,2]]
print(data)
target = df.iloc[:,-1]
print(target)
```

```python
import matplotlib.pyplot as plt

#อยากได้กราฟ Bar = plt.bar

plt.plot(['2565','2566','2567'],[10,40,30])
plt.axis([0,2,0,100])
plt.ylabel('percent')
plt.xlabel('Number')
plt.show()
```
