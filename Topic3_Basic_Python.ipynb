**Problem: Python Lists**

```python
#เก็บคะแนนนักศึกษา
listScore = [50, 65, 90, 75, 80, 80, 85, 90]
print(listScore)

#หาคะแนนต่ำสุด
min(listScore)
print(min(listScore))

#ลบคะแนนต่ำสุด
listScore.remove(min(listScore))
print(listScore)
```

**Problem: Python Tuples**

```python
#เก็บคะแนน
tupleScore = (50, 65, 90, 75, 80, 80, 85, 90, 99, 95)
print(tupleScore)

#หา Max
max(tupleScore)
print(max(tupleScore))

#หา Min
min(tupleScore)
print(min(tupleScore))

#แสดงผล
print(max(tupleScore) - min(tupleScore))
```
