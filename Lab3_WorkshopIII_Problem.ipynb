* สามารถสร้างแบบจำลองโดยใช้เทคนิค KNeighborsClassifier และข้อมูลทั้งหมด และทดสอบ
แบบจำลองโดยใช้ข้อมูลทั้งหมดได้ (Training set = testset) และแสดงข้อมูลผลลัพธ์ (y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
* สามารถแสดง Confusion Matrix ออกทางจอภาพได้
* สามารถประเมินแบบจำลองโดยใช้ Accuracy, Precision, Recall, F-measure ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ

```python
from sklearn import datasets
from sklearn import neighbors
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

#นำเข้าข้อมูล iris.data และ iris.target จาก sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target

print("X = ",X)
print("\n")

#สร้าง Model KNeighborsClassifier
model = neighbors.KNeighborsClassifier()
model.fit(X, y)


#แสดงผล Y_TRUE และ Y_Predict
y_true = iris.target
knn= neighbors.KNeighborsClassifier()
knn.fit(X, y_true)
y_pred = knn.predict(X)

print('Y_TRUE =', y_true)
print('Y_Predict =', y_pred)
print("\n")


#แสดงผล Confusion_matrix
conf_matrix = confusion_matrix(y_true, y_pred)
print(conf_matrix)
print("\n")

#แสดงผลการประเมินผลโมเดล ในส่วนของ Accuracy, Precision, Recall และ F-measure
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print("Accuracy:" ,'%.2f' %(accuracy) )
print("Precisio:" ,'%.2f' %(precision) )
print("Recall:" ,'%.2f' %(recall) )
print("F-measure" ,'%.2f' %(f1) )

```

เพราะมันจะ random ค่าทุกครั้งที่รันโปรแกรม เครื่องนักศึกษารัน 2 ครั้งยังไม่เหมือนกัน

```python
import numpy as np
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score, KFold
import matplotlib.pyplot as plt

#นำเข้าข้อมูล iris จาก sklearn
iris = datasets.load_iris()
X = iris.data
y = iris.target


#สามารถสร้างแบบจำลองโดยใช้เทคนิค Artificial Neural Network และชุดข้อมูลที่แบง 8 ชุดนั้น (hint: MLPClassifier(hidden_layer_sizes=(10,10,10), max_iter=1000))
kf = KFold(n_splits=8, shuffle=True, random_state=1)
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)

#สามารถประเมินแบบจำลองโดยใช้ Accuracy ได้ โดยแสดงค่าของแต่ละรอบ และค่าเฉลี่ยทั้งหมด ออกทางจอภาพ
scores = cross_val_score(mlp, X, y, cv=kf, scoring="accuracy")
print(scores)
print(scores.mean())


#สามารถแสดงกราฟเส้น (Line Graph) ของค่า Accuracy ทั้ง 8 ออกทางจอภาพ
plt.plot(scores)
plt.xlabel("Fold")
plt.ylabel("Accuracy")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from pandas.core.common import random_state
from sklearn.datasets import load_iris
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import neighbors
from sklearn import tree
from sklearn import metrics

#สามารถนำเข้าข้อมูล iris data จาก sklearn datasets ได้ถูกต้อง
iris = load_iris()
X = iris.data
y = iris.target

#และแบ่งข้อมูล Training set 80% และ Testset 20%
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=1)


#โดยแสดงข้อมูล x_train, y_train, x_test, y_test ออกทางจอภาพได้
print("x_train =", pd.DataFrame(x_train[:10]))
print("x_test = ", pd.DataFrame(x_test[:10]))

print("y_train = ",y_train)
print("y_test = ",y_test)


#สามารถสร้างแบบจำลองโดยใช้ 2 เทคนิค คือ KNeighborsClassifier, Decision Tree ความลึกสูงสุดของต้นไม้คือ 4
knn = KNeighborsClassifier()
dt = DecisionTreeClassifier(max_depth=4, random_state=1)

# เทรนโมเดล
knn.fit(x_train, y_train)
dt.fit(x_train, y_train)

# ทำนายผล
y_pred_knn = knn.predict(x_test)
y_pred_dt = dt.predict(x_test)


#แสดงข้อมูลผลลัพธ์ (y_true) และข้อมูลที่ได้จากการทำนาย (y_pred) ออกทางจอภาพได้
print("Y_TRUE (y_test) =", y_test)
print("Y_Predict (y_predict) (KN) =", y_pred_knn)

print("Y_Predict (y_predict) (Decision Tree) =", y_pred_dt)


#สามารถแสดง Confusion Matrix ออกทางจอภาพได้
cm_knn = confusion_matrix(y_test, y_pred_knn)
cm_dt = confusion_matrix(y_test, y_pred_dt)


#สามารถประเมินแบบจำลองโดยใช้ F-measure ได้ โดยแสดงค่าทั้งหมดออกทางจอภาพ
f_knn = f1_score(y_test, y_pred_knn, average='weighted')
f_dt = f1_score(y_test, y_pred_dt, average='weighted')

results = pd.DataFrame({
    'Model': ['KNN', 'Decision Tree'],
    'F-measure': [f_knn, f_dt]
})


#สามารถแสดงตารางเปรียบเทียบผลการประเมินของทุกเทคนิคได้
print("\nConfusion Matrix (KNN)\n", cm_knn)
print("\nConfusion Matrix (Decision Tree)\n", cm_dt)
print("\nผลการเปรียบเทียบโมเดล")
print(results)
```
